# Automated Excel Sales Report

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("superstore_sales.xlsx")
```

# Data Overview

```python
print("\n First 5 rows: ")
df.head()
```

```python
print("\n Last 5 rows: ")
df.tail()
```

```python
df.info()
```

# Data Quality validation

```python
 df.isnull().sum() #check for missing values
```

```python
df.duplicated().sum() #check for duplicated rows
```

```python
df['Order Date']= pd.to_datetime(df['Order Date']) #convert order date to date time.
df['Order Date']
```

```python
df['Month']= df['Order Date'].dt.to_period('M') #create month column
df['Month']
```

# Monthly Sales Analysis

```python
monthly_sales= df.groupby('Month')['Sales'].sum()
monthly_sales= monthly_sales.round(0).astype(int) # converts float to int, so that data looks professional and clean when printed.
monthly_sales
```

```python
monthly_sales.plot(kind='line', figsize=(14,6), marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

### Insights:
- Sales trend over time
- Identify peak and low months
- Seasonal buying behavior

# Category wise Sales

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales = category_sales.round(0).astype(int)
category_sales
```

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.bar(category_sales.index,category_sales.values,color='red')
ax.set_title("Sales by Category")
ax.set_xlabel("Category")
ax.set_ylabel("Total Sales")  
plt.show()
```

### Insights
- Highest sales were in Technology and lowest in Furniture.
- Category contribution to total revenue

# Top 10 Products by Sales

```python
top_products= df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products= top_products.round(0).astype(int)
top_products
```

```python
fig, ax= plt.subplots(figsize=(10,6))
ax.barh(top_products.index,top_products.values,color='green')
ax.set_title("Top 10 Products by Sales")
ax.set_ylabel("Product Name")
ax.set_xlabel("Total Sales")
ax.invert_yaxis() # highest on top
plt.show()        
```

### Insights:
- Canon imageCLASS 2200 Advanced Copier was sold the most.
- High Speed Automatic Electric Letter Opener was sold least in Top 10.  
- Inventory & marketing focus

# Region wise Sales

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales = region_sales.round(0).astype(int)
region_sales
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(region_sales.index, region_sales.values, color='purple')
ax.set_title("Sales by Region")
ax.set_xlabel("Region")
ax.set_ylabel("Total Sales")
plt.show()
```

### Insights
- Best performing region is West and least South.
- Market expansion opportunities

# Category wise Profit

```python
category_profit = df.groupby('Category')['Profit'].sum()
category_profit = category_profit.round(0).astype(int)
category_profit
```

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.bar(category_profit.index,category_profit.values,color='blue')
ax.set_title("Profit by Category")
ax.set_xlabel("Category")
ax.set_ylabel("Total Profit")  
plt.show()
```

### Insights
- Most profitable category is Technology and least Furniture.
- Cost or pricing issues

# Overall Profit Margin

```python
overall_margin = (df['Profit'].sum() / df['Sales'].sum()) * 100
overall_margin = round(overall_margin, 2)
print(f"Overall Profit Margin: {overall_margin}%")
```

# Business Insights

## 1. Sales Trend Over Time

- Monthly sales show high volatility, with multiple sharp spikes and drops.
- Overall trend is upward, especially from 2016 to 2017, indicating business growth.
- Sales peaks suggest seasonal demand, likely driven by promotions, year-end buying, or bulk orders.

###     Business Impact

- Company should plan inventory & staffing around peak months.
- Forecasting models can reduce stockouts during high-demand periods.

## 2. Sales by Category

- Technology is the highest revenue-generating category.
- Office Supplies contributes strong, stable sales.
- Furniture generates the lowest sales among the three.

### Business Impact

- Technology should remain the primary growth focus.
- Furniture may need better pricing, bundling, or marketing strategy.
- Office Supplies can be leveraged for consistent cash flow.

## 3. Sales by Region

- West region generates the highest total sales.
- East region performs strongly and is close behind.
- Central region is moderate.
- South region is the lowest-performing.

### Business Impact

- Replicate West region strategies (pricing, promotions, sales channels) in other regions.
- South region needs targeted campaigns or distribution improvements.
- Consider reallocating marketing budget toward high-ROI regions.

## 4. Profit by Category

- Technology is the most profitable category.
- Office Supplies has strong profitability.
- Furniture shows very low profit, despite sales.

### Business Impact

- Furniture likely suffers from high costs, discounts, or logistics expenses.
- Immediate review needed on cost structure and supplier pricing for Furniture.
- Focus growth efforts on high-margin Technology products.

## 5. Top 10 Products by Sales

- Sales are concentrated in a few high-performing products.
- The top product significantly outperforms others.
- Many top products belong to Technology and Office Equipment.

### Business Impact

- Protect top products with inventory priority and supplier contracts.
- Upsell and cross-sell related products.
- Reduce dependency risk by developing mid-tier products.

## 6. Overall Profitability

- Overall Profit Margin: 12.47%

### Interpretation

- Margin is healthy but has room for improvement.
- Improving low-margin categories (Furniture) could significantly increase total profit.
- Small pricing or cost optimizations can have large impact at scale.

## 7. Key Strategic Recommendations

- Double down on Technology (highest sales + highest profit).
- Fix or rethink Furniture (low profit warning sign).
- Strengthen South region with targeted actions.
-  Use seasonality insights for better forecasting.
- Optimize pricing to push profit margin beyond 15%

# Automated Export of Cleaned Data & Business Insights to Excel

```python
output_file= "clean_business_report.xlsx"

insights = [
    "Technology is the most profitable and highest-revenue category.",
    "Furniture shows low profit, indicating potential cost or pricing inefficiencies.",
    "West and East regions contribute the highest share of total sales.",
    "Monthly sales trends reveal seasonal demand spikes.",
    f"Overall profit margin is {overall_margin}%, indicating moderate operational efficiency."
]

insights_df = pd.DataFrame({
    "Business Insights": insights
})

monthly_sales.index = monthly_sales.index.strftime("%b-%Y")

with pd.ExcelWriter(output_file) as writer:
    df.to_excel(writer, sheet_name= "Cleaned_Data", index=False)
    monthly_sales.to_excel(writer, sheet_name="Monthly_Sales")
    category_sales.to_excel(writer, sheet_name="Category_Sales")
    region_sales.to_excel(writer, sheet_name="Region_Sales")
    category_profit.to_excel(writer, sheet_name="Category_Profit")

    summary_df = pd.DataFrame({
        "Metric": ["Overall Profit Margin (%)"],
        "Value": [overall_margin]
    })
    summary_df.to_excel(writer, sheet_name="Business_Summary", index=False)

    insights_df.to_excel(writer, sheet_name="Business_Insights", index=False)

print("Excel file 'cleaned_business_report.xlsx' created successfully in the same folder.")
```
